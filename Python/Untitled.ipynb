```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("daily.csv")
```

```python
def plot_data(data, state_name="NY"):
    ny_data = data[data["state"]==state_name][["date", "positiveIncrease", "totalTestResultsIncrease"]].dropna()
    ny_data["positiveRate"] = ny_data["positiveIncrease"] / ny_data["totalTestResultsIncrease"]
    import matplotlib.pyplot as plt
    plt.plot([i for i in range(len(ny_data["positiveRate"]))], ny_data["positiveIncrease"][::-1])
    plt.plot([i for i in range(len(ny_data["positiveRate"]))], ny_data["totalTestResultsIncrease"][::-1])
    plt.show()
    plt.plot([i for i in range(len(ny_data["positiveRate"]))], ny_data["positiveRate"][::-1])
```

```python
plot_data(data)
```

```python
plot_data(data, "IL")
```

```python
plot_data(data, "CA")
```
